## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Set index to mouse ID . FZ
mouse_metadata.set_index('Mouse ID')
study_results.set_index('Mouse ID')

# Combine the data into a single dataset
combined_Dset = pd.merge(study_results, mouse_metadata, right_index=False, left_index=False)

# Display the data table for preview

combined_Dset
```

```python
# Checking the number of mice
num_of_mice = combined_Dset.nunique()
num_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicateRowsDF = combined_Dset[combined_Dset.duplicated()]
duplicateRowsDF
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_Dset_no_duplicates = combined_Dset.drop_duplicates()
```

```python
# Checking the number of mice in the clean DataFrame.
combined_Dset_no_duplicates
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

temp_tumor_count = combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].count()
temp_tumor_count.rename(columns = {'Tumor Volume (mm3)':'Number of Measurements'}, inplace = True)
stat_summary = temp_tumor_count
stat_summary['Tumor Volume Mean'] = round(combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].mean(),2)
stat_summary['Tumor Volume Median'] = round(combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].median(),2)
stat_summary['Tumor Volume Varience'] = round(combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].var(),2)
stat_summary['Tumor Volume Standard Deviation'] = round(combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].std(),2)
stat_summary['Tumor Volume SEM'] = round(combined_Dset_no_duplicates.groupby('Drug Regimen')[['Tumor Volume (mm3)']].sem(),2)
stat_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

four_regimens = combined_Dset_no_duplicates.loc[combined_Dset_no_duplicates['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol','Ceftamin'])]

four_regimens
# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = four_regimens.groupby('Mouse ID').last()
last_timepoint

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
# no need to do this as I already have the tumor volume in my current df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
df_Infubinol = last_timepoint.loc[last_timepoint['Drug Regimen'] == 'Infubinol', 'Tumor Volume (mm3)'].to_frame()
df_Ceftamin = last_timepoint.loc[last_timepoint['Drug Regimen'] == 'Ceftamin', 'Tumor Volume (mm3)'].to_frame()
df_Ramicane = last_timepoint.loc[last_timepoint['Drug Regimen'] == 'Ramicane', 'Tumor Volume (mm3)'].to_frame()
df_Capomulin = last_timepoint.loc[last_timepoint['Drug Regimen'] == 'Capomulin', 'Tumor Volume (mm3)'].to_frame()

# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
   # Locate the rows which contain mice on each drug and get the tumor volumes
   # add subset 
   # Determine outliers using upper and lower bounds

############Infubinol###############
quartiles = df_Infubinol.quantile([.25,.5,.75])
list_quartiles = quartiles['Tumor Volume (mm3)'].tolist()
list_quartiles
lowerq = list_quartiles[0]
upperq = list_quartiles[2]
iqr = upperq-lowerq

print(f"The lower quartile of Infubinol is: {lowerq}")
print(f"The upper quartile of Infubinol is: {upperq}")
print(f"The interquartile range of Infubinol is: {iqr}")
print(f"The the median of Infubinol is: {list_quartiles[1]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
print('\n')
######################################


############Ceftamin###############
quartiles = df_Ceftamin.quantile([.25,.5,.75])
list_quartiles = quartiles['Tumor Volume (mm3)'].tolist()
list_quartiles
lowerq = list_quartiles[0]
upperq = list_quartiles[2]
iqr = upperq-lowerq

print(f"The lower quartile of Ceftamin is: {lowerq}")
print(f"The upper quartile of Ceftamin is: {upperq}")
print(f"The interquartile range of Ceftamin is: {iqr}")
print(f"The the median of Ceftamin is: {list_quartiles[1]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
print('\n')
######################################

############Ramicane###############
quartiles = df_Ramicane.quantile([.25,.5,.75])
list_quartiles = quartiles['Tumor Volume (mm3)'].tolist()
list_quartiles
lowerq = list_quartiles[0]
upperq = list_quartiles[2]
iqr = upperq-lowerq

print(f"The lower quartile of Ramicane is: {lowerq}")
print(f"The upper quartile of Ramicane is: {upperq}")
print(f"The interquartile range of Ramicane is: {iqr}")
print(f"The the median of Ramicane is: {list_quartiles[1]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
print('\n')
######################################

############Capomulin###############
quartiles = df_Capomulin.quantile([.25,.5,.75])
list_quartiles = quartiles['Tumor Volume (mm3)'].tolist()
list_quartiles
lowerq = list_quartiles[0]
upperq = list_quartiles[2]
iqr = upperq-lowerq

print(f"The lower quartile of Capomulin is: {lowerq}")
print(f"The upper quartile of Capomulin is: {upperq}")
print(f"The interquartile range of Capomulin is: {iqr}")
print(f"The the median of Capomulin is: {list_quartiles[1]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
print('\n')
######################################
```
